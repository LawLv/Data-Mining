```python
# 第一题
import pandas as pd

# 加载数据集
data = pd.read_csv("HW2_1_data.csv")

# 找出包含缺失值的样本
missing_samples = data[data.isnull().any(axis=1)]

# 删除包含缺失值的样本
data_cleaned = data.dropna()

# 确认删除后的数据集
print("删除前数据集大小:", data.shape)
print("删除后数据集大小:", data_cleaned.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
import time

# 计算代码运行时间
start_time = time.time()

# 加载数据集
data = data_cleaned

# 划分特征和标签
X = data.drop(columns=['quality'])
y = data['quality']

# 数据标准化
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=13)

# 初始化并训练随机森林分类器
rf_clf = RandomForestClassifier()
rf_clf.fit(X_train, y_train)

# 初始化并训练梯度提升树分类器
gbt_clf = GradientBoostingClassifier()
gbt_clf.fit(X_train, y_train)

# 在测试集上评估随机森林分类器
rf_pred = rf_clf.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_pred)

# 在测试集上评估梯度提升树分类器
gbt_pred = gbt_clf.predict(X_test)
gbt_accuracy = accuracy_score(y_test, gbt_pred)

end_time = time.time()
execution_time = end_time - start_time

print("Random Forest Accuracy:", rf_accuracy)
print("Gradient Boosting Trees Accuracy:", gbt_accuracy)
print("Execution Time:", execution_time, "seconds")
# 总时间小于2min,所以每种分类器均少于2min
```

```python
# 第二题
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import time

# 加载数据集
data = pd.read_csv("HW2_2_data.csv")

# 划分特征和标签
X = data.drop(columns=['class'])
y = data['class']

# 数据标准化
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=8)

# 初始化随机森林分类器
rf_clf = RandomForestClassifier()

# 计算代码运行时间
start_time = time.time()

# 训练随机森林分类器
rf_clf.fit(X_train, y_train)

# 在测试集上评估随机森林分类器
rf_pred = rf_clf.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_pred)

end_time = time.time()
execution_time = end_time - start_time

print("Random Forest Accuracy:", rf_accuracy)
print("Execution Time:", execution_time, "seconds")
```
